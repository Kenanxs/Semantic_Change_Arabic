```python
# Print the preprocessed text in chunks
chunk_size = 1000  # number of characters per chunk
lines_to_print = 5  # number of lines to print

start = 0
end = chunk_size
line_count = 0

while start < len(preprocessed_text) and line_count < lines_to_print:
    print(preprocessed_text[start:end])
    start += chunk_size
    end += chunk_size
    line_count += 1
```

```python
# preprocessing files in T1 + saving to new output directory to use later
import os
from tqdm import tqdm

corpus_dir = "/content/drive/My Drive/OpenITI/T1"
output_dir = "/content/drive/My Drive/OpenITI/T1_processed_arabertv2"

# Count the number of text files to process
total_files = 0

for root, dirs, files in os.walk(corpus_dir):
    for file in files:
        file_path = os.path.join(root, file)
        if not file.endswith((".yml", ".md")):
            total_files += 1

with tqdm(total=total_files, desc="Processing files", unit="file") as pbar:
    for root, dirs, files in os.walk(corpus_dir):
        for file in files:
            file_path = os.path.join(root, file)

            # Exclude metadata files and readme files
            if not file.endswith((".yml", ".md")):
                # Check if the output file already exists
                output_file_path = os.path.join(output_dir, file)
                if os.path.exists(output_file_path):
                    pbar.write(f"Skipping {file} as it has already been processed.")
                    pbar.update(1)
                    continue

                # Read the file content
                with open(file_path, "r", encoding="utf-8") as f:
                    file_content = f.read()

                # Preprocess the file content
                preprocessed_text = arabert_prep.preprocess(file_content)

                # Save the preprocessed text to a new file
                with open(output_file_path, "w", encoding="utf-8") as f:
                    f.write(preprocessed_text)

                pbar.update(1)
```

```python
# just a check for files to make sure theyre not empty or very short
# Get a list of all files in the directory
import os
output_dir = "/content/drive/My Drive/OpenITI/T1_processed_arabertv2"

# Specifying the minimum content length threshold
minimum_content_length = 100

# Iterate through the files
for file_name in os.listdir(output_dir):
    # Open the file
    file_path = os.path.join(output_dir, file_name)
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()

    # Check if the file is empty or its content is too short
    if len(content) < minimum_content_length:
        print(f"File: {file_name} - Content length: {len(content)}")
```

```python
def get_directory_char_count(directory_path):
    total_char_count = 0

    for root, dirs, files in os.walk(directory_path):
        for file in files:
            file_path = os.path.join(root, file)

            # Exclude metadata files and readme files
            if not file.endswith((".yml", ".md")):
                with open(file_path, "r", encoding="utf-8") as f:
                    content = f.read()
                    char_count = len(content)
                    total_char_count += char_count

    return total_char_count

# Directory paths
directory1_path = "/content/drive/My Drive/OpenITI/T1_processed_arabertv2"
directory2_path = "/content/drive/My Drive/OpenITI/T2_processed_arabertv2"

# Calculate character counts
directory1_char_count = get_directory_char_count(directory1_path)
directory2_char_count = get_directory_char_count(directory2_path)

# Print the character counts
print("T1 Character Count:", directory1_char_count)
print("T2 Character Count:", directory2_char_count)
```
